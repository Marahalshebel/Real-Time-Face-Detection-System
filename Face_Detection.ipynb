# Real-Time Face Detection using OpenCV# Real-Time Face Detection using OpenCV

This project detects human faces in real time using a webcam and OpenCV's Haar Cascade classifier.

```python
import cv2
import time
from datetime import datetime
```

```python
cap = cv2.VideoCapture(0)
```

```python
face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)
```

```python
start_time = time.time()

while True:
    ret, frame = cap.read()

    if not ret:
        break

    # Convert frame to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces
    faces = face_cascade.detectMultiScale(
        gray,
        scaleFactor=1.1,
        minNeighbors=5
    )

    face_count = len(faces)

    # Show message if no face is detected
    if face_count == 0:
        cv2.putText(
            frame,
            "No Face Detected",
            (20, 120),
            cv2.FONT_HERSHEY_COMPLEX,
            0.8,
            (0, 0, 255),
            2
        )

    # Draw corner rectangles around detected faces
    for (x, y, w, h) in faces:

        line = 25
        thickness = 2

        # Top Left
        cv2.line(frame, (x, y), (x + line, y), (0, 255, 0), thickness)
        cv2.line(frame, (x, y), (x, y + line), (0, 255, 0), thickness)

        # Top Right
        cv2.line(frame, (x + w, y), (x + w - line, y), (0, 255, 0), thickness)
        cv2.line(frame, (x + w, y), (x + w, y + line), (0, 255, 0), thickness)

        # Bottom Left
        cv2.line(frame, (x, y + h), (x + line, y + h), (0, 255, 0), thickness)
        cv2.line(frame, (x, y + h), (x, y + h - line), (0, 255, 0), thickness)

        # Bottom Right
        cv2.line(frame, (x + w, y + h), (x + w - line, y + h), (0, 255, 0), thickness)
        cv2.line(frame, (x + w, y + h), (x + w, y + h - line), (0, 255, 0), thickness)

        cv2.putText(
            frame,
            "Detected Face",
            (x, y - 10),
            cv2.FONT_HERSHEY_TRIPLEX,
            0.8,
            (0, 255, 0),
            2
        )

    # Display number of detected faces
    cv2.putText(
        frame,
        f"Faces Detected: {face_count}",
        (20, 40),
        cv2.FONT_HERSHEY_TRIPLEX,
        1,
        (0, 0, 255),
        2
    )

    # Display current date and time
    current_time = datetime.now().strftime("%d-%m-%Y %H:%M:%S")

    cv2.putText(
        frame,
        current_time,
        (20, 80),
        cv2.FONT_HERSHEY_DUPLEX,
        0.7,
        (0, 0, 255),
        2
    )

    # Welcome message (shown for the first 3 seconds)
    if time.time() - start_time < 3:
        cv2.putText(
            frame,
            "Welcome to Face Detection System",
            (40, 160),
            cv2.FONT_HERSHEY_COMPLEX,
            0.8,
            (255, 255, 255),
            2
        )

    # Show camera window
    cv2.imshow("Real-Time Face Detection", frame)

    key = cv2.waitKey(1) & 0xFF

    # Save image when S is pressed
    if key == ord('s'):
        filename = datetime.now().strftime("Face_%Y%m%d_%H%M%S.jpg")
        cv2.imwrite(filename, frame)
        print(f"Image Saved: {filename}")

    # Quit when Q is pressed
    if key == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
